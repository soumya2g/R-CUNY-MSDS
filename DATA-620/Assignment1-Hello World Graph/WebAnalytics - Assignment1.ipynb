### Krackhardt Kite Graph:

In graph theory, the Krackhardt kite graph is a simple graph with ten nodes. The graph is named after David Krackhardt, a researcher of social network theory.

Krackhardt introduced the graph in 1990 to distinguish different concepts of centrality. It has the property that the vertex with maximum degree (labeled 3 in the figure, with degree 6), the vertex with maximum betweenness centrality (labeled 7), and the two vertices with maximum closeness centrality (labeled 5 and 6) are all different from each other.

**Reference:** https://en.wikipedia.org/wiki/Krackhardt_kite_graph

### Generate a simple Krackhardt Kite graph with Python Function

```python
import matplotlib.pyplot as plt
import networkx as nx

G = nx.krackhardt_kite_graph()

nx.draw(G, with_labels=True)
plt.show()
```

### Generate Krackhardt Kite graph from scratch

**Step1:** Create an empty graph with no nodes and no edges.

```python
G=nx.Graph()
```

**Step2:** To start simple, I have a added a single node to the graph G using *add_node()* function.

```python
G.add_node('Andre')

nx.draw(G,with_labels=True )
plt.show()
```

**Step3:** But instead of adding one node at a time, I can add the remaining nodes at once using *add_nodes_from()* function.

```python
members = ['Beverly','Ed','Carol','Diane','Fernando','Ed','Garth','Heather','Ike','Jane']
G.add_nodes_from(members)

nx.draw(G,with_labels=True )
plt.show()
```

**Step4:** Add an edge to the graph using *add_edge()*.

```python
G.add_edge('Andre','Beverly')

nx.draw(G,with_labels=True )
plt.show()
```

**Step5:** Similar to nodes, instead of adding one edge at a time, we can add rest of the edges to the graph at once using *add_edges_from()*.

```python
G.add_edges_from([('Andre','Carol'),('Andre','Diane'),('Fernando','Andre'), ('Beverly','Ed'),('Beverly','Diane'),('Beverly','Garth'),
                  ('Ed','Diane'),('Ed','Garth'),('Garth','Diane'),('Garth','Fernando'),('Garth','Heather'),
                  ('Heather','Fernando'),('Heather','Ike'),('Ike','Jane'),('Fernando','Diane'),('Fernando','Carol'),
                 ('Carol','Diane')])

plt.figure(figsize=(8,8))
nx.draw(G,node_size=400,with_labels=True )
plt.title("Krackhardt Kite Graph", fontsize=20)
plt.show()
```
